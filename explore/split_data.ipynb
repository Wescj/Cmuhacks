```python
import pandas as pd
import os
import numpy as np
```

```python
# Path ke file CSV dan folder output
input_file = "./data/analyst_ratings_filtered.csv"
output_folder = "./split_analyst_ratings"

# Pastikan folder tujuan ada
os.makedirs(output_folder, exist_ok=True)

# Baca file
df = pd.read_csv(input_file)

# Loop berdasarkan unique stock
for stock, group in df.groupby("stock"):
    output_path = os.path.join(output_folder, f"{stock}.csv")
    group.to_csv(output_path, index=False)
    print(f"Saved {output_path}")

print("Splitting done ")
```

```python
data_df = pd.read_csv("./data/stock_data/symbols_valid_meta.csv")
data_df.head()
```

```python
data_df["Market Category"].unique()
```

```python
data_df["Market Category"].rename({" ": "CONSUMER", "S": "FINANCE", "Q": "TECHNOLGY", "G": "HEALTHCARE"}, inplace=True)
data_df["Market Category"].head()
```

```python
# clean + relabel the codes
cat_map = {
    "": "CONSUMER",        # blank after strip()
    "S": "FINANCE",
    "Q": "TECHNOLOGY",     # note the spelling
    "G": "HEALTHCARE"
}

s = data_df["Market Category"].astype(str).str.strip()   # remove spaces
data_df["Market Category"] = s.replace(cat_map)

# quick sanity check
print(data_df["Market Category"].value_counts(dropna=False))
print(data_df["Market Category"].head())
```

```python
data_df.head()
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Define probability for ESG = 1 (mean 0.5, sd 0.15 clipped between 0 and 1)
# You can adjust mean and sd
market_cats = data_df["Market Category"].unique()

esg_probs = {}
for cat in market_cats:
    prob = np.clip(np.random.normal(loc=0.5, scale=0.15), 0, 1)
    esg_probs[cat] = prob

print("Assigned ESG probabilities per category:", esg_probs)

# Assign ESG using the probabilities per Market Category
data_df["ESG"] = data_df["Market Category"].apply(
    lambda x: np.random.choice([0, 1], p=[1-esg_probs[x], esg_probs[x]])
)

# Check distribution
print(data_df.groupby("Market Category")["ESG"].mean())
```
